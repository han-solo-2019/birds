# Results

The file contains results for different architectures of CNN's tested. Presented Classification Reports were calculated on a test set (10% of a whole dataset). 
* Maximum number of classes tested is 19
* Each CNN is trained on 5-sec Melspectrograms filtered with high-pass filters
* Classes are highly imbalanced: smallest class have 74 samples in the test set, whreaes the biggest have 3741
