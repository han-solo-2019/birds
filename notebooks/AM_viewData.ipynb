```python
# find all of the files in the directory
import os


birds=[] # list of all birds (Ember, Phyll...)
blist=[] # list of files for one bird
flist=[] # list of all files for all birds. A list of blists. 
for root, dirs, files in os.walk("../data/mels-19class/"):
    if root == "../data/mels-19class/":
        birds=dirs        

for bird in birds:
    for root, dirs, files in os.walk("../data/mels-19class/"+bird):
        for file in files:
            if file.endswith(".png"):
                blist.append(os.path.join(root, file))
    print("Found ", len(blist), ' files for ', bird)
    flist.append(blist)
    blist=[]
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

for i in range(5):
    img=mpimg.imread(flist[i][100]) # show image from class i with number 500
    imgplot = plt.imshow(img)
    plt.show()
```
